```python
from hdfs import InsecureClient
from io import BytesIO

user = "hadoop"
host = "http://namenode:9870"

data_list = []
for i in range(16):
    path = f"/user/hadoop/지역_위치별(주유소){i}.xls"
    hdfs = InsecureClient(host, user=user)
    with hdfs.read(path) as reader:
        file_content = BytesIO(reader.read())  # Read into a BytesIO buffer
        data_list.append(pd.read_excel(file_content, engine="xlrd", header=2))
data_list[2]
```

```python
total_data = pd.concat([data_list[0],data_list[1],data_list[2],data_list[3],data_list[4],data_list[5],data_list[6],data_list[7],data_list[8],data_list[9],data_list[10],data_list[11],data_list[12],data_list[13],data_list[14],data_list[15]],ignore_index=True)
total_data['휘발유'] = total_data['휘발유'].replace('-',0).astype(int)
# total_data['경유'] = total_data['경유'].replace('-',0).astype(int)
gas_filtered_data = total_data.loc[total_data['휘발유'] > 0]
# diesel_filtered_data = total_data.loc[total_data['경유'] > 0]
gas_min5_data = gas_filtered_data.sort_values(by='휘발유',ascending=True)
gas_max5_data = gas_filtered_data.sort_values(by='휘발유',ascending=False)
# diesel_min5_data = diesel_filtered_data.sort_values(by='경유',ascending=True)

min_data = gas_min5_data.head(5)
max_data = gas_max5_data.head(5)
min_data = min_data.values.tolist()
max_data = max_data.values.tolist()
min_data
max_data

min_data[0].extend([35.2074400081561,129.071593475684])
min_data[1].extend([35.2064419619732,129.073915104508])
min_data[2].extend([35.2057636764632,129.073888079651])
min_data[3].extend([35.2025841983223,129.099112586558])
min_data[4].extend([35.2067375098508,129.102787528496])
max_data[0].extend([35.1123617607372,129.038760123867])
max_data[1].extend([35.0894084391224,128.897442858073])
max_data[2].extend([35.0822718307859,129.012778541809])
max_data[3].extend([35.0742699304025,129.006501809393])
max_data[4].extend([35.0949319380558,129.047732982229])
```

```python
import folium

min_df = pd.DataFrame(min_data)
min_df.columns = ['지역','상호','주소','상표','전화번호','셀프여부','고급휘발유','휘발유','경유','실내등유','위도','경도']
min_df
max_df = pd.DataFrame(max_data)
max_df.columns = ['지역','상호','주소','상표','전화번호','셀프여부','고급휘발유','휘발유','경유','실내등유','위도','경도']

coordinates_min = min_df[['위도','경도']].to_numpy()
coordinates_max = max_df[['위도','경도']].to_numpy()
# print(coordinates_min)

center_lat, center_long = (min_df['위도'].astype(float).sum() + max_df['위도'].astype(float).sum()) / 10, (min_df['경도'].astype(float).sum() + max_df['경도'].astype(float).sum()) / 10

m = folium.Map(location=[center_lat, center_long],zoom_start=11)

for i in range(len(coordinates_min)):
    lat, long = coordinates_min[i]
    folium.CircleMarker(
        location=[lat, long],
        radius = 10,
        color = 'blue',
        fill = True,
        fill_color = 'white',
        fill_opacity = 0.5,
    ).add_to(m)

for i in range(len(coordinates_max)):
    lat, long = coordinates_max[i]
    folium.CircleMarker(
        location=[lat, long],
        radius = 10,
        color = 'red',
        fill = True,
        fill_color = 'black',
        fill_opacity = 0.5,
    ).add_to(m)
m
```
