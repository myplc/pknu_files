```python
from kafka import KafkaProducer
import json
import time

producer = KafkaProducer(
    bootstrap_servers = 'kafka:9092',
    value_serializer = lambda v:json.dumps(v).encode('utf-8')
)
num = 0
how = 1000
delay = 1

def generate_smartcar_data(num):
    import random
    return{
        "car_id":f"{num}/{how}",
        "speed": round(random.uniform(80,100), 2),
        "latitude": round(random.uniform(34,36), 2),
        "longitude": round(random.uniform(124,126), 2),
        "timestamp": time.time()
    }

if __name__ == "__main__":
    try:
        for i, r in enumerate(range(how),1):
            data = generate_smartcar_data(i)
            producer.send("smart-car", value = data)
            print(f"Send:{data}")
            time.sleep(delay)
    except:
        print("Error")
    finally:
        producer.close()
```

```python
from kafka import KafkaConsumer
import json

# Kafka  
consumer = KafkaConsumer(
    'smart-car',  # 
    bootstrap_servers='kafka:9092',  # Kafka  
    auto_offset_reset='latest',  #  (earliest: , latest:  )
    enable_auto_commit=True,  #    
    group_id=None,  #   
    value_deserializer=lambda x: json.loads(x.decode('utf-8')),  #   
    consumer_timeout_ms=100000  #    
)

#    
for message in consumer:
    print(message.value)
```
